```python
import random
#サイコロ3つ（3色1個ずつ）を同時に投げた場合
class Dice:
    def __init__(self, color):
        self.color = color  # サイコロの色
        self.result_list = []  # 振った結果を格納するリスト
        self.result_sum = 0  # 合計値

    def roll_dice(self):
        # サイコロを1回振り、1から6までのランダムな整数を返す
        return random.randint(1, 6)

    def calculate_sum(self):
        # リストの合計値を計算し、result_sumに格納する
        self.result_sum = sum(self.result_list)

# 各色のサイコロをインスタンス化
red_dice = Dice("赤")
blue_dice = Dice("青")
yellow_dice = Dice("黄")

print("--- 3色のサイコロを同時に転がします！ ---")

# 15回同時にサイコロを転がす
for i in range(15):
    red_roll = red_dice.roll_dice()
    blue_roll = blue_dice.roll_dice()
    yellow_roll = yellow_dice.roll_dice()

    red_dice.result_list.append(red_roll)
    blue_dice.result_list.append(blue_roll)
    yellow_dice.result_list.append(yellow_roll)

    print(f"{i + 1}回目: 赤={red_roll}, 青={blue_roll}, 黄={yellow_roll}")

# 15回したから、各サイコロの合計値を計算してください。
red_dice.calculate_sum()
blue_dice.calculate_sum()
yellow_dice.calculate_sum()

# 各色の合計値（計算結果）を表示
print("\n--- 各色のサイコロの最終合計値 ---")
print(f"赤色のサイコロ合計: {red_dice.result_sum}")
print(f"青色のサイコロ合計: {blue_dice.result_sum}")
print(f"黄色のサイコロ合計: {yellow_dice.result_sum}")

#if-elif を使って一番大きい値を見つける
if red_dice.result_sum >= blue_dice.result_sum and \
   red_dice.result_sum >= yellow_dice.result_sum:
    winning_color = red_dice.color
    winning_sum = red_dice.result_sum
elif blue_dice.result_sum >= red_dice.result_sum and \
     blue_dice.result_sum >= yellow_dice.result_sum:
    winning_color = blue_dice.color
    winning_sum = blue_dice.result_sum
else: # yellow_dice.result_sum が一番大きい場合（または引き分けの場合も含む）
    winning_color = yellow_dice.color
    winning_sum = yellow_dice.result_sum

print(f"\n一番値が大きかったのは【{winning_color}】のサイコロで、合計は【{winning_sum}】でした！")
```

```python
import random
#サイコロ3つ（3色1個ずつ）を同時に投げた場合
class Dice:
    def __init__(self, color):
        self.color = color  # サイコロの色
        self.result_list = []  # 振った結果を格納するリスト
        self.result_sum = 0  # 合計値

    def roll_dice(self):
        # サイコロを1回振り、1から6までのランダムな整数を返す
        return random.randint(1, 6)

    def calculate_sum(self):
        # リストの合計値を計算し、result_sumに格納する
        self.result_sum = sum(self.result_list)

# 各色のサイコロをインスタンス化
red_dice = Dice("赤")
blue_dice = Dice("青")
yellow_dice = Dice("黄")

print("--- 3色のサイコロを同時に転がします！ ---")

# 15回同時にサイコロを転がす
for i in range(15):
    red_roll = red_dice.roll_dice()
    blue_roll = blue_dice.roll_dice()
    yellow_roll = yellow_dice.roll_dice()

    red_dice.result_list.append(red_roll)
    blue_dice.result_list.append(blue_roll)
    yellow_dice.result_list.append(yellow_roll)

    print(f"{i + 1}回目: 赤={red_roll}, 青={blue_roll}, 黄={yellow_roll}")

# 各サイコロの合計値を計算
red_dice.calculate_sum()
blue_dice.calculate_sum()
yellow_dice.calculate_sum()

# 各色の合計値を表示
print("\n--- 各色のサイコロの最終合計値 ---")
print(f"赤色のサイコロ合計: {red_dice.result_sum}")
print(f"青色のサイコロ合計: {blue_dice.result_sum}")
print(f"黄色のサイコロ合計: {yellow_dice.result_sum}")

# 最も合計値が高かった色を特定
dice_results = [
    (red_dice.result_sum, red_dice.color),
    (blue_dice.result_sum, blue_dice.color),
    (yellow_dice.result_sum, yellow_dice.color)
]
#こっちのほうがシンプルなコード
winning_result = max(dice_results, key=lambda item: item[0])
winning_sum = winning_result[0]
winning_color = winning_result[1]

print(f"\n一番値が大きかったのは【{winning_color}】のサイコロで、合計は【{winning_sum}】でした！")
```
